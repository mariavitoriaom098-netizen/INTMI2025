##decimal de PI

```python
from decimal import Decimal, getcontext

def calcular_pi(n):
    # Define a precisão (algumas casas extras para evitar erro)
    getcontext().prec = n + 5

    # Algoritmo de Gauss–Legendre
    a = Decimal(1)
    b = Decimal(1) / Decimal(2).sqrt()
    t = Decimal(1) / Decimal(4)
    p = Decimal(1)

    for _ in range(20):
        an = (a + b) / 2
        b = (a * b).sqrt()
        t -= p * (a - an) ** 2
        a = an
        p *= 2

    pi = (a + b) ** 2 / (4 * t)
    getcontext().prec = n
    return +pi  # aplica arredondamento

# Entrada do usuário
n = int(input("Quantas casas decimais de π você quer calcular? "))
pi = calcular_pi(n)
print(f"\nπ com {n} casas decimais:\n{pi}")
```

##Crie um algoritmo que leia a sua idade e mostre as frases na tela:

Até 12 anos – Você é uma criança!

13 – 18 – Você é adolescente!

19 – 60 – Você é adulto!

Acima de 60 – Você é idoso!

```python
 idade = int(input("Digite a sua idade:"))
if idade <= 12:
    print("Você é uma criança!")
elif idade >= 13 and idade <= 18:
    print("Você é um adolescente!")
elif idade >= 19 and idade <= 60:
    print("Você é um adulto!")
else:
    print("Você é um idoso!")
```

##Crie um algoritimo que receba 2 números e uma escolha que mostre o sequinte menu:

selecione o sinal:

1- Adição

2- Subtração

3- Multiplicação

4- Divisão


caso não escolha um número válido apareça na tela: Opção invalida!

se escolher 1, some dois números e mostre o resultado

se escolher 2, subtraia os dois números e mostre o resultado

e assim com as outras opções

```python
# Entrada dos números
num1 = float(input("Digite o primeiro número:"))
num2 = float(input("Digite o segundo número:"))

# Menu de opções
print("Selecione o sinal:")
print("1 - Adição")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")

# Escolha do usuário
opcao = int(input("Digite a opção desejada (1 a 4): "))

# Estrutura
if opcao == 1:
    resultado = num1 + num2
    print("Resultado da adição: {resultado}")
elif opcao == 2:
    resultado = num1 - num2
    print("Resultado da subtração: {resultado}")
elif opcao == 3:
    resultado = num1 * num2
    print("Resultado da multiplicação: {resultado}")
elif opcao == 4:
    # Evita divisão por zero
    if num2 != 0:
        resultado = num1 / num2
        print("Resultado da divisão: {resultado}")
    else:
        print("Erro! Não é possível dividir por zero.")
else:
    print("Opção inválida!")
```
